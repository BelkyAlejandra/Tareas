# DBSCAN

```python
# Surpress any warnings:
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
import string

import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

import seaborn as sns

sns.set_context('notebook')
sns.set_style('white')

# Import matplotlib for 3d plotting:
import matplotlib.pyplot as plt

# Make matplotlib work in jupyter notebook
%matplotlib inline
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-ML0187EN-SkillsNetwork/labs/module%202/data/DBSCAN_exercises.csv')
df.head()
```

```python
plt.scatter(df['x'], df['y'],color='y')
plt.show()
```

# 1

```python
cluster = DBSCAN(eps=2, min_samples=10)
cluster.fit(df)
print(f'DBSCAN found {len(set(cluster.labels_) - set([-1]))} clusters.')
```

# 2

```python
print(f'DBSCAN encontr que  {(cluster.labels_ == -1).sum()} puntos son ruido')
#print(f'{(1000*3)/100}')
```

```python
df.head()
```

```python
df1=df.copy()
```

```python
clusters = cluster.fit_predict(df1)
df1["label"] = clusters
```
